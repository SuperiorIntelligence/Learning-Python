### ➕ All Arithmetic Operators in Python

```python
a = 17
b = 5

print(f"a = {a}, b = {b}")
print(f"\na + b  = {a + b}   (Addition)")
print(f"a - b  = {a - b}   (Subtraction)")
print(f"a * b  = {a * b}   (Multiplication)")
print(f"a / b  = {a / b}   (Division — always returns float)")
print(f"a // b = {a // b}   (Floor Division — discards decimal)")
print(f"a % b  = {a % b}   (Modulo — remainder of division)")
print(f"a ** b = {a ** b}  (Exponentiation — a to the power of b)")
```

```python
# Division: regular vs floor division
print("--- Division Types ---")
print(f"10 / 3  = {10 / 3}      (regular: always a float, keeps decimals)")
print(f"10 // 3 = {10 // 3}        (floor: rounds DOWN to nearest whole)")
print(f"10 % 3  = {10 % 3}        (modulo: only the remainder)")

print("\n--- Negative floor division ---")
print(f"-10 // 3 = {-10 // 3}       (rounds DOWN, i.e. more negative)")
print(f"10 // -3 = {10 // -3}       (same rule applies)")
```

```python
# Practical use of modulo (%)
number = 47

is_even = (number % 2 == 0)   # even numbers have no remainder when divided by 2
is_odd  = (number % 2 != 0)

print(f"{number} % 2 = {number % 2}")
print(f"Is {number} even? {is_even}")
print(f"Is {number} odd?  {is_odd}")
```

```python
# Exponentiation
base   = 2
power  = 10
result = base ** power
print(f"{base} ** {power} = {result}")

# Square root using exponent (power of 0.5)
number = 144
sqrt   = number ** 0.5
print(f"Square root of {number} = {sqrt}")
```

### 📊 Augmented Assignment Operators

```python
# Augmented assignment: shorthand for updating a variable
score = 100
print(f"Initial score: {score}")

score += 10     # same as: score = score + 10
print(f"After += 10: {score}")

score -= 5      # same as: score = score - 5
print(f"After -= 5:  {score}")

score *= 2      # same as: score = score * 2
print(f"After *= 2:  {score}")

score //= 3     # same as: score = score // 3
print(f"After //=3:  {score}")
```

### 🏆 Operator Precedence — Which Runs First?

Python evaluates expressions in a specific order, just like math (PEMDAS/BODMAS).

```python
# Operator Precedence in Python (from HIGHEST to LOWEST)
# 1. ()              Parentheses
# 2. **              Exponentiation (right to left)
# 3. *, /, //, %     Multiplication and division (left to right)
# 4. +, -            Addition and subtraction (left to right)
# 5. Comparisons     ==, !=, >, <, >=, <=
# 6. not             Logical NOT
# 7. and             Logical AND
# 8. or              Logical OR

print("=== WITHOUT parentheses (default precedence) ===")

# Multiplication runs before addition
print(f"2 + 3 * 4 = {2 + 3 * 4}")  

# Exponentiation runs before subtraction
print(f"10 - 2 ** 3 = {10 - 2 ** 3}")  

# Division runs before addition
print(f"15 / 3 + 2 = {15 / 3 + 2}")  

# Exponentiation is right-to-left
print(f"2 ** 3 ** 2 = {2 ** 3 ** 2}")  # 2 ** (3 ** 2) = 2 ** 9 = 512


print("\n=== WITH parentheses (you control the order) ===")

# Parentheses force addition first
print(f"(2 + 3) * 4 = {(2 + 3) * 4}")  

# Parentheses force subtraction first
print(f"(10 - 2) ** 3 = {(10 - 2) ** 3}")  

# Change order in division and addition
print(f"15 / (3 + 2) = {15 / (3 + 2)}")  


print("\n=== COMPARISON and LOGICAL operators ===")

# Comparison runs before logical AND
print(f"5 > 3 and 2 < 1 = {5 > 3 and 2 < 1}")  

# NOT has higher precedence than AND
print(f"not True and False = {not True and False}")  

# Parentheses change logical evaluation
print(f"not (True and False) = {not (True and False)}")
```

```python
# Mixed arithmetic and comparison operators
x = 5
y = 3

# Arithmetic is evaluated BEFORE comparison
result = x + y > 2 * y     # becomes: (5+3) > (2*3) → 8 > 6 → True
print(f"x + y > 2 * y  →  {x} + {y} > 2 * {y}  →  {x+y} > {2*y}  →  {result}")

result2 = x ** 2 == y ** 2 + 16   # becomes: 25 == 9 + 16 → 25 == 25 → True
print(f"x**2 == y**2 + 16  →  {x**2} == {y**2} + 16  →  {x**2} == {y**2+16}  →  {result2}")
```

We introduced `and`, `or`, `not` in Session 2. Now let's understand their **precedence** and how they work with arithmetic and comparison operators in one expression.

```python
# Full precedence order (high → low):
#  (**) > (*, /, //, %) > (+, -) > (==, !=, >, <, >=, <=) > (not) > (and) > (or)

age   = 20
score = 75
name  = "Alice"

# Python evaluates this step by step:
# 1. Arithmetic:   score + 5  → 80
# 2. Comparisons:  age >= 18  → True  |  80 >= 60  → True
# 3. Logical and:  True and True → True
result = age >= 18 and score + 5 >= 60
print(f"age >= 18 and score + 5 >= 60  →  {result}")

# Another example:
# 1. Arithmetic: age * 2 → 40,  score - 10 → 65
# 2. Comparisons: 40 > 35 → True | 65 < 60 → False
# 3. not False → True
# 4. True and True → True
result2 = age * 2 > 35 and not score - 10 < 60
print(f"age*2 > 35 and not score-10 < 60  →  {result2}")
```

```python
# not has higher precedence than and, which is higher than or
a = True
b = False
c = True

print("--- Precedence: not > and > or ---")
print(f"a or b and c         = {a or b and c}     (b and c first → False; then a or False → True)")
print(f"not a or b           = {not a or b}   (not a first → False; then False or b → False)")
print(f"not (a or b)         = {not (a or b)}   (a or b first → True; then not True → False)")
print(f"a and not b or c     = {a and not b or c}     (not b=True; a and True=True; True or c=True)")
```

```python
# Complete precedence demonstration — building from known concepts
password_length = 10
has_uppercase   = True
has_digit       = True
is_banned       = False

is_valid = password_length >= 8 and has_uppercase and has_digit and not is_banned

# Step by step:
# 1. password_length >= 8     → True
# 2. True and has_uppercase   → True
# 3. True and has_digit       → True
# 4. not is_banned            → True
# 5. True and True            → True

print(f"Password valid: {is_valid}")
```

### 🧠 The `if` Statement — Run code only when a condition is True

```python
# Basic if statement
temperature = 35

if temperature > 30:
    print("It is hot outside!")       # only runs if condition is True
    print("Remember to drink water.") # still inside the if block

print("Program continues here regardless.")  # always runs
```

### 🔀 `if` / `else` — Two possible paths

```python
# if / else — exactly one of the two blocks runs
score = 72

if score >= 60:
    print(f"Score: {score} → PASSED ✅")
else:
    print(f"Score: {score} → FAILED ❌")
```

### 🔀 `if` / `elif` / `else` — Multiple possible paths

```python
# if / elif / else — checks conditions in order, runs the FIRST one that is True
score = 85

if score >= 90:
    grade = "A"
    remark = "Excellent!"
elif score >= 80:
    grade = "B"
    remark = "Very good!"
elif score >= 70:
    grade = "C"
    remark = "Good."
elif score >= 60:
    grade = "D"
    remark = "Needs improvement."
else:
    grade = "F"
    remark = "Failed."

print(f"Score : {score}")
print(f"Grade : {grade}")
print(f"Remark: {remark}")
```

```python
# ⚠️ Important: elif stops checking as soon as one is True!
score = 95

if score >= 60:        # True → runs this, then SKIPS all remaining elif/else
    print("Passed")   # ← this runs
elif score >= 80:
    print("Good")     # ← this is NEVER reached even though 95 >= 80 is True!
elif score >= 90:
    print("Excellent")# ← also never reached

# Lesson: Order your conditions from MOST SPECIFIC to LEAST SPECIFIC!
```

```python
# Correct order: most specific first!
score = 95

if score >= 90:         # most specific (highest threshold) first
    print("Excellent")  # ← this correctly runs now
elif score >= 80:
    print("Good")
elif score >= 60:
    print("Passed")
else:
    print("Failed")
```

```python
# Real-world example: Simple calculator using if/elif/else
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if operator == "+":
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")
elif operator == "-":
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")
elif operator == "*":
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
        print(f"{num1} / {num2} = {result}")
    else:
        print("Error: Cannot divide by zero!")
else:
    print(f"Unknown operator: '{operator}'")
```

A **nested condition** is an `if` statement placed inside another `if` block. Each level adds one more layer of indentation.

```python
# Basic nesting: check one condition, then another inside it
age    = 22
has_id = True

if age >= 18:
    print("You are old enough.")          # outer if — runs when age >= 18
    if has_id:
        print("You may enter. ✅")        # inner if — runs only when BOTH conditions are met
    else:
        print("Please show your ID. 🪪") # inner else — age ok but no ID
else:
    print("You are too young to enter. ❌") # outer else — age < 18
```

```python
# A more detailed example: Online store discount calculator
# Uses only: variables, input, f-strings, comparison, arithmetic

is_member     = True
purchase_total = 150.0

if is_member:
    print("Welcome back, member!")
    if purchase_total >= 200:
        discount_rate = 0.20    # 20% for members with big orders
        discount_label = "20% (member + large order)"
    elif purchase_total >= 100:
        discount_rate = 0.10    # 10% for members with medium orders
        discount_label = "10% (member + medium order)"
    else:
        discount_rate = 0.05    # 5% for members with small orders
        discount_label = "5% (member)"
else:
    print("Hello, guest!")
    if purchase_total >= 200:
        discount_rate = 0.05    # 5% for non-members with big orders
        discount_label = "5% (large order)"
    else:
        discount_rate = 0.0     # no discount for guests with small orders
        discount_label = "No discount"

discount_amount = purchase_total * discount_rate
final_price     = purchase_total - discount_amount

print(f"\nPurchase total : ${purchase_total:.2f}")
print(f"Discount       : {discount_label}")
print(f"Discount amount: ${discount_amount:.2f}")
print(f"Final price    : ${final_price:.2f}")
```

```python
# ⚠️ Avoid deep nesting when possible — use 'and' to combine conditions

# Version 1: Nested (harder to read)
age = 22
has_id = True
is_banned = False

if age >= 18:
    if has_id:
        if not is_banned:
            print("[Nested] Access granted ✅")

# Version 2: Flat with 'and' (cleaner, easier to read)
if age >= 18 and has_id and not is_banned:
    print("[Flat]   Access granted ✅")

# Both do the same thing — prefer the flat version when possible!
```

### 🔢 `int` — Integer (whole numbers)

```python
# int stores whole numbers — no decimal point
age         = 25
year        = 2025
temperature = -8
big_number  = 1_000_000   # underscores make large numbers readable!

print(f"age         = {age}         type: {type(age).__name__}")
print(f"year        = {year}        type: {type(year).__name__}")
print(f"temperature = {temperature}          type: {type(temperature).__name__}")
print(f"big_number  = {big_number}    type: {type(big_number).__name__}")

# Arithmetic on ints
print(f"\nage + 5    = {age + 5}")
print(f"year - 100 = {year - 100}")
print(f"10 / 3     = {10 / 3}   ← dividing two ints gives a FLOAT!")
print(f"10 // 3    = {10 // 3}   ← floor division keeps it as int")
```

### 🔣 `float` — Floating-Point (decimal numbers)

```python
price      = 19.99
pi         = 3.14159
scientific = 1.5e3    # scientific notation: 1.5 × 10³ = 1500.0
small      = 2.5e-4   # 2.5 × 10⁻⁴ = 0.00025

print(f"price      = {price}        type: {type(price).__name__}")
print(f"pi         = {pi}      type: {type(pi).__name__}")
print(f"scientific = {scientific}       type: {type(scientific).__name__}")
print(f"small      = {small}      type: {type(small).__name__}")

# ⚠️ Floating point precision — a known quirk
print(f"\n0.1 + 0.2 = {0.1 + 0.2}   ← not exactly 0.3! (floating point limitation)")
print(f"round(0.1 + 0.2, 2) = {round(0.123 + 0.2232, 2)}   ← use round() for display")      #You can think of round() as: "Round a number to the nearest value with the precision I specify."

#Important: Python's "Banker's Rounding"
# Python uses round half to even.
print(round(2.5))
print(round(3.5))
```

### 🔤 `str` — String (text)

```python
# Strings can use single, double, or triple quotes
name1 = 'Alice'            # single quotes
name2 = "Bob"              # double quotes — same result
message = "It's a great day!"   # single quote inside double is fine
quote = 'He said "hello"'       # double quote inside single is fine
multiline = """Line 1
Line 2
Line 3"""

print(name1, name2)
print(message)
print(quote)
print(multiline)
```

Strings in Python have built-in methods — functions that you call **on** a string using dot notation: `string.method()`.

### 🔡 Case Methods

```python
text = "hello world, python is amazing!"

print(f"Original  : {text}")
print(f".upper()  : {text.upper()}")
print(f".lower()  : {text.lower()}")
print(f".title()  : {text.title()}")                    # str.title() capitalizes the first character of every word in the string
print(f".capitalize(): {text.capitalize()}")            # str.capitalize() only capitalizes the first character of the entire string
```

```python
# Practical use: case-insensitive comparison
user_input = input("Type 'yes' to continue: ")

if user_input.lower() == "yes":
    print("Continuing... ✅")
else:
    print("Stopping. ❌")

# This works whether the user types: YES, Yes, yEs, yes, etc.
```

### ✂️ Strip Methods — Remove Unwanted Spaces

```python
raw = "   hello world   "

print(f"Original  : '{raw}'")
print(f".strip()  : '{raw.strip()}'")
print(f".lstrip() : '{raw.lstrip()}'")
print(f".rstrip() : '{raw.rstrip()}'")

# Practical use: clean up user input before processing
username = input("Enter username: ").strip()   # removes accidental spaces
print(f"Username (cleaned): '{username}'")
print(f"Length: {len(username)} characters")
```

### 🔍 Search Methods

```python
sentence = "Python programming is fun and Python is powerful"

# .find() — returns the index of first match, or -1 if not found
position = sentence.find("Python")
print(f".find('Python')       = {position}    (found at index {position})")
print(f".find('Java')         = {sentence.find('Java')}    (not found → -1)")

# .count() — how many times does a substring appear?
count = sentence.count("Python")
print(f"\n.count('Python')      = {count}    ('Python' appears {count} times)")
print(f".count('is')           = {sentence.count('is')}    ('is' appears {sentence.count('is')} times)")

# .startswith() and .endswith() — True/False checks
print(f"\n.startswith('Python') = {sentence.startswith('Python')}")
print(f".endswith('powerful')  = {sentence.endswith('powerful')}")
print(f".endswith('fun')       = {sentence.endswith('fun')}")
```

### 🔄 Replace Method

```python
message = "Hello, my name is Alice. Alice loves coding."

# .replace(old, new) — replaces ALL occurrences by default
new_message = message.replace("Alice", "Bob")
print(f"Original: {message}")
print(f"Replaced: {new_message}")

# .replace(old, new, count) — replace only first N occurrences
partial = message.replace("Alice", "Bob", 1)   # replace only the first one
print(f"Partial : {partial}")

# ⚠️ Strings are immutable — replace creates a NEW string, doesn't change the original!
print(f"Original (unchanged): {message}")
```

### 📦 Split

```python
# .split() — breaks a string into a list of substrings
sentence = "Python is easy to learn"
words = sentence.split()          # splits on whitespace by default
print(f"Original: '{sentence}'")
print(f".split() : {words}")
print(f"Number of words: {len(words)}")

# Split on a specific character
csv_line = "Alice,25,Amsterdam,Python"
fields = csv_line.split(",")      # split on comma
print(f"\nCSV line: '{csv_line}'")
print(f".split(','): {fields}")
```

### 🔲 Slicing — Extract Part of a String

```python
# Slicing syntax: string[start:end:step]
# start is INCLUSIVE, end is EXCLUSIVE
text = "Python Programming"
#        0123456789...

print(f"text             = '{text}'")
print(f"text[0:6]        = '{text[0:6]}'    (chars 0 to 5)")
print(f"text[7:]         = '{text[7:]}'  (from index 7 to end)")
print(f"text[:6]         = '{text[:6]}'    (from start to index 5)")
print(f"text[-11:]       = '{text[-11:]}'  (last 11 characters)")
print(f"text[::2]        = '{text[::2]}'    (every 2nd character)")
print(f"text[::-1]       = '{text[::-1]}'  (reversed!)")
```

  <h2 style="color: #e94560; margin-top: 0;">🎯 Exercise 1 — Smart Score Grader</h2>

  <p><strong>Grading scale:</strong></p>
  <ul style="line-height: 2;">
    <li>90–100 → A — "Outstanding performance!"</li>
    <li>80–89  → B — "Great job!"</li>
    <li>70–79  → C — "Good work."</li>
    <li>60–69  → D — "You passed, but review the material."</li>
    <li>0–59   → F — "Study harder next time."</li>

```python
# Exercise 1: Smart Score Grader

name  = input("Enter your name: ").strip().title()
score = int(input("Enter your exam score (0-100): "))

if score < 0 or score > 100:
    print("Invalid score! Please enter a value between 0 and 100.")
else:
    if score >= 90:
        grade   = "A"
        message = "Outstanding performance!"
    elif score >= 80:
        grade   = "B"
        message = "Great job!"
    elif score >= 70:
        grade   = "C"
        message = "Good work."
    elif score >= 60:
        grade   = "D"
        message = "You passed, but review the material."
    else:
        grade   = "F"
        message = "Study harder next time."

    print(f"\n🎓 Result for {name}")
    print("-" * 30)
    print(f"Score  : {score} / 100")
    print(f"Grade  : {grade}")
    print(f"Remark : {message}")
```

    <li>Total number of characters (including space)</li>
    <li>Number of characters without spaces</li>
    <li>Whether the name starts with a vowel (a)</li>
    <li>The first name and last name separately (assume one space between them)</li>
  </ul>
  <p><em>Concepts used: input(), strip(), upper(), lower(), title(), len(), replace(), split(), startswith(), f-string, if/elif/else</em></p>

```python
# Exercise 3: Name Analyzer

full_name = input("Enter your full name (first and last): ").strip()

# Extract parts
parts      = full_name.split()
first_name = parts[0] if len(parts) >= 1 else full_name
last_name  = parts[1] if len(parts) >= 2 else "(not provided)"

# Calculations using Session 2 concepts: len, replace
total_chars    = len(full_name)
chars_no_space = len(full_name.replace(" ", ""))

# Does the name start with the letter a?
if first_name.lower()[0] == "a" :
    starts_with = True
else :
   starts_with = False 

print(f"\n📛 Name Analysis")
print("=" * 35)
print(f"Original         : {full_name}")
print(f"UPPER            : {full_name.upper()}")
print(f"lower            : {full_name.lower()}")
print(f"Title            : {full_name.title()}")
print(f"Total characters : {total_chars}")
print(f"Without spaces   : {chars_no_space}")
print(f"First name       : {first_name}")
print(f"Last name        : {last_name}")

if starts_with:
    print(f"✅ Starts with  a")
else:
    print(f"❌ Doesn't Starts with a")
```

<div style="color: #000; background: #fffbeb; border-radius: 12px; padding: 25px; margin: 15px 0; border: 2px solid #f6ad55;">
  <h2 style="color: #744210; margin-top: 0;">🎯 Exercise 4 — Trip Cost Calculator</h2>
  <p><strong>Task:</strong> Calculate the cost of a road trip using arithmetic operators.</p>
  <p><strong>Given information (ask with input):</strong></p>
  <ul style="line-height: 2;">
    <li>Distance (km)</li>
    <li>Fuel consumption (liters per 100 km)</li>
    <li>Fuel price (per liter)</li>
    <li>Number of passengers (to split cost)</li>
  </ul>
  <p><strong>Calculate and display:</strong> total fuel needed, total fuel cost, cost per person. Then use if/elif/else to print whether the cost per person is cheap (&lt;20), moderate (20–50), or expensive (&gt;50).</p>
  <p><em>Concepts used: input(), float(), int(), arithmetic operators, f-string (with .2f), comparison operators, if/elif/else</em></p>
</div>

```python
# Exercise 4: Trip Cost Calculator

print("🚗 Road Trip Cost Calculator")
print("-" * 30)

distance        = float(input("Distance (km): "))
fuel_per_100km  = float(input("Fuel consumption (L/100km): "))
price_per_liter = float(input("Fuel price per liter ($): "))
passengers      = int(input("Number of passengers: "))

# Arithmetic calculations
total_fuel_liters = (distance / 100) * fuel_per_100km
total_cost        = total_fuel_liters * price_per_liter
cost_per_person   = total_cost / passengers

# Display results using f-strings with .2f formatting
print(f"\n📊 Trip Summary")
print("=" * 30)
print(f"Distance         : {distance} km")
print(f"Fuel needed      : {total_fuel_liters:.2f} liters")
print(f"Total fuel cost  : ${total_cost:.2f}")
print(f"Passengers       : {passengers}")
print(f"Cost per person  : ${cost_per_person:.2f}")

# Classify the cost
if cost_per_person < 20:
    verdict = "Cheap 💚 — great deal!"
elif cost_per_person <= 50:
    verdict = "Moderate 🟡 — reasonable."
else:
    verdict = "Expensive 🔴 — consider alternatives."

print(f"Cost level       : {verdict}")
```
