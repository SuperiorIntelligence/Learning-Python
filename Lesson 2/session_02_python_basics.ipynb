### 🖨️ The `print()` Function

The `print()` function displays output to the screen. It's the first function every Python developer learns!

```python
# Your very first Python statement!
print("Hello, World!")
```

```python
# print() can display different types of data
print("I am learning Python")       # string
print(42)                            # integer number
print(3.14)                          # float (decimal) number
print(True)                          # boolean
print("Name:", "Alice", "Age:", 25)  # multiple values separated by comma
```

```python
# print() options: sep and end
print("Python", "is", "awesome", sep="-")       # custom separator
print("First line", end=" ")                    # don't go to new line
print("same line")                              # continues on same line
print("")                                          # empty line
```

### 🎤 Getting Input from the User — `input()`

The `input()` function **pauses** the program and waits for the user to type something. It always returns a **string**.

```python
# Basic input example
name = input("What is your name? ")
print("Hello,", name)
```

```python
# IMPORTANT: input() always returns a STRING
# If you need a number, you must convert it!
age_text = input("How old are you? ")
age = int(age_text)          # Convert string → integer
print("In 10 years you will be:", age + 10)
```

```python
# Shorter way — convert directly inside input()
age = int(input("Enter your age: "))
print("Next year you will be:", age + 1)
```

### 🎨 f-strings — String Formatting Made Easy

f-strings (formatted string literals) let you **embed variables directly inside strings**. They are the most modern and readable way to format strings in Python 3.6+.

```python
# Old way (messy!)
name = "Alice"
age = 25
print("My name is " + name + " and I am " + str(age) + " years old.")

# Modern way — f-string (clean and readable!)
print(f"My name is {name} and I am {age} years old.")
```

```python
# f-strings can contain expressions (calculations) inside the braces!
price = 49.99
quantity = 3

print(f"Price per item: ${price}")
print(f"Total for {quantity} items: ${price * quantity}")
print(f"After 10% discount: ${price * quantity * 0.9:.2f}")  # .2f = 2 decimal places
```

```python
# f-string formatting options
pi = 3.14159265358979

print(f"Default:       {pi}")
print(f"2 decimals:    {pi:.2f}")
print(f"5 decimals:    {pi:.5f}")
print(f"As percentage: {0.875:.1%}")    # .1% → percentage format with 1 decimal place (0.875 → 87.5%)
print(f"Width 10:      {pi:10.2f}")     # 10.2f → width=10 chars, 2 decimals, float format (right-aligned)
```

```python
# Complete example combining input() and f-string
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")
birth_year = int(input("Enter your birth year: "))

current_year = 2025
age = current_year - birth_year

print(f"\n--- Profile ---")
print(f"Full Name : {first_name} {last_name}")
print(f"Birth Year: {birth_year}")
print(f"Age       : {age} years old")
```

### 📐 Python is Block-Based — Indentation MATTERS!

Unlike many languages that use `{}` braces, **Python uses indentation (spaces) to define code blocks**. This is one of Python's most distinctive features.

```python
# Example: if statement with proper indentation
temperature = 35

if temperature > 30:
    print("It's hot outside!")       # indented = inside the if block
    print("Drink water 💧")          # still inside the if block

print("This always runs")            # NOT indented = outside the if block
```

```python
# ❌ IndentationError — Python will refuse to run this!
# temperature = 35
# if temperature > 30:
# print("Missing indent!")   # ERROR: Expected an indented block

# ✅ Correct:
temperature = 35
if temperature > 30:
    print("Correct indentation!")   # 4 spaces (standard)
```

  <ul style="color: #1a1313;line-height: 2;">
    <li>Use <strong>4 spaces</strong> per indent level (Python community standard — PEP 8)</li>
    <li>NEVER mix tabs and spaces — pick one and stick to it</li>
    <li>Every line in the same block must have the same indentation</li>

### 🔠 Case Sensitivity — Python Cares About Capital Letters!

```python
# Python is CASE SENSITIVE
Name = "Alice"    # different variables!
name = "Bob"
NAME = "Charlie"

print(Name)   # Alice
print(name)   # Bob
print(NAME)   # Charlie

# Functions are also case sensitive
print("hello")   # ✅ works
# Print("hello") # ❌ NameError: name 'Print' is not defined
```

### 🏷️ Variable Naming Rules and Conventions

```python
# ✅ VALID variable names
user_name = "Alice"          # snake_case — Python convention!
firstName = "Bob"            # camelCase — common in other languages
age = 25
_private_var = "hidden"      # starts with underscore = convention for "private"
counter1 = 0                 # can include numbers (but not at start!)
MAX_SIZE = 100               # ALL_CAPS = convention for constants

print(user_name, firstName, age)
```

```python
# ❌ INVALID variable names — these will cause errors!
# 1name = "bad"       # SyntaxError: cannot start with a digit
# my-name = "bad"     # SyntaxError: hyphens not allowed
# my name = "bad"     # SyntaxError: spaces not allowed
# class = "bad"       # SyntaxError: 'class' is a reserved keyword

# Python reserved keywords (cannot be used as variable names):
import keyword
print("Python keywords:", keyword.kwlist)
```

```python
# Best Practice: Use descriptive snake_case names

# ❌ Bad names (hard to understand)
x = 25
y = "Alice"
z = 1750.00

# ✅ Good names (self-documenting)
user_age = 25
customer_name = "Alice"
monthly_salary = 1750.00

print(f"{customer_name} is {user_age} years old and earns ${monthly_salary}/month")
```

### 🗒️ Single-Line Comments — `#`

Comments are notes for humans. Python **ignores** everything after `#` on a line.

```python
# This is a single-line comment — Python ignores this

name = "Alice"   # This is an inline comment

# Comments help explain WHY, not just WHAT
tax_rate = 0.21   # VAT rate in the Netherlands (2025)

price = 100
total = price * (1 + tax_rate)   # Add VAT to get final price
print(f"Price with VAT: €{total:.2f}")
```

### 📚 Multi-Line Docstrings — `"""..."""`

Docstrings are used to **document** functions, classes, and modules. They explain **what** something does.

```python
# A simple variable with a docstring

message = "Hello, Alice! Welcome to Python class! 🎉"

"""
This is a docstring example.

It explains what this part of the code does:
- Stores a greeting message in a variable
- Can be used for documentation purposes
"""

print(message)

# Accessing the docstring
print("\n--- Docstring ---")
print(__doc__)
```

```python
# Multi-line strings (not docstrings) — useful for long text
# These are just triple-quoted strings used as values
poem = """
Roses are red,
Python is great,
Code every day,
Don't hesitate!
"""
print(poem)
```

### 🗃️ What is a Variable?

A variable is a **named container** that stores a value in memory. Think of it like a labeled box.

```python
# Assigning variables
city = "Amsterdam"       # str  — text
population = 921000      # int  — whole number
temperature = 12.5       # float — decimal number
is_capital = False       # bool — True or False

print(city, population, temperature, is_capital)
```

### 🔍 The `type()` Function — What Type is This Variable?

```python
# type() tells you the data type of any value
print(type("Hello"))       # <class 'str'>
print(type(42))            # <class 'int'>
print(type(3.14))          # <class 'float'>
print(type(True))          # <class 'bool'>
print(type([1, 2, 3]))     # <class 'list'>
```

```python
# Checking types of variables
name = "Alice"
score = 98.5
passed = True

print(f"name  = {name!r}  → type: {type(name).__name__}")            # !r → uses repr() to show the exact value (with quotes), helpful for debugging
print(f"score = {score}     → type: {type(score).__name__}")
print(f"passed= {passed}    → type: {type(passed).__name__}")
```

### 📏 The `len()` Function — How Long is This?

```python
# len() returns the number of characters in a string (or items in a list)
greeting = "Hello, World!"
print(f"String: '{greeting}'")
print(f"Length: {len(greeting)} characters")

# len() on different types
word = "Python"
sentence = "I love programming"

print(f"\n'{word}' has {len(word)} characters")
print(f"'{sentence}' has {len(sentence)} characters")
```

### 🔄 Type Conversion — Converting Between Data Types

```python
# String to Number conversion
age_str = "25"             # This is a string
age_int = int(age_str)     # Convert to integer
age_flt = float(age_str)   # Convert to float

print(f"'{age_str}' → int: {age_int}, type: {type(age_int).__name__}")
print(f"'{age_str}' → float: {age_flt}, type: {type(age_flt).__name__}")
```

```python
# Number to String conversion
score = 95
score_str = str(score)     # Convert to string
print(f"{score} → '{score_str}', type: {type(score_str).__name__}")

# WHY convert to string? To concatenate with +
message = "Your score is: " + str(score) + " points"
print(message)
```

```python
# Float to Integer (loses the decimal part!)
price = 29.99
rounded_price = int(price)   # truncates — does NOT round!
print(f"{price} → {rounded_price}  (decimal part is DROPPED, not rounded)")

# To properly round, use round()
print(f"round({price}) = {round(price)}")
```

```python
# Boolean conversions
print(bool(0))       # False  — zero is falsy
print(bool(1))       # True   — any non-zero number is truthy
print(bool(""))      # False  — empty string is falsy
print(bool("hi"))    # True   — non-empty string is truthy
print(int(True))     # 1
print(int(False))    # 0
```

Comparison operators **compare two values** and always return `True` or `False`.

```python
a = 10
b = 20

print(f"a = {a}, b = {b}")
print(f"\na == b  (equal to)              → {a == b}")
print(f"a != b  (not equal to)          → {a != b}")
print(f"a >  b  (greater than)          → {a > b}")
print(f"a <  b  (less than)             → {a < b}")
print(f"a >= b  (greater than or equal) → {a >= b}")
print(f"a <= b  (less than or equal)    → {a <= b}")
```

```python
# Real-world example: Comparing user input
user_age = int(input("Enter your age: "))
minimum_age = 18

is_adult = user_age >= minimum_age
print(f"\nYour age: {user_age}")
print(f"Minimum age: {minimum_age}")
print(f"Are you an adult? {is_adult}")
```

```python
# Comparing strings
password_correct = "secret123"
user_input = "secret123"

print(f"Password match: {user_input == password_correct}")

# ⚠️ String comparison is case-sensitive!
print(f"'Python' == 'python' → {'Python' == 'python'}")  # False!
print(f"'Python' == 'Python' → {'Python' == 'Python'}")  # True
```

```python
# ⚠️ IMPORTANT: == vs = (a very common mistake!)
x = 5         # ASSIGNMENT: store value 5 in x
print(x == 5) # COMPARISON: is x equal to 5? → True
print(x == 10)# COMPARISON: is x equal to 10? → False

# Remember:
# =   assigns a value
# ==  checks if two values are equal
```

Logical operators **combine multiple comparisons**. Python uses `and`, `or`, `not` (unlike many languages that use `&&`, `||`, `!`).

### `and` — Both conditions must be True

```python
# and: returns True ONLY if BOTH sides are True
print(True  and True)   # True  ✅ both true
print(True  and False)  # False ❌ right side is false
print(False and True)   # False ❌ left side is false
print(False and False)  # False ❌ both false
```

```python
# Real example: Can this person enter a club?
age = 22
has_id = True

can_enter = (age >= 18) and has_id
print(f"Age: {age}, Has ID: {has_id}")
print(f"Can enter the club: {can_enter}")
```

### `or` — At least one condition must be True

```python
# or: returns True if AT LEAST ONE side is True
print(True  or True)    # True  ✅
print(True  or False)   # True  ✅ left side is enough
print(False or True)    # True  ✅ right side is enough
print(False or False)   # False ❌ both are false
```

```python
# Real example: Is today a day off?
day = "Saturday"
is_holiday = False

is_day_off = (day == "Saturday") or (day == "Sunday") or is_holiday
print(f"Today is {day}, Holiday: {is_holiday}")
print(f"Is it a day off? {is_day_off}")
```

### `not` — Reverses the Boolean value

```python
# not: flips True to False and False to True
print(not True)    # False
print(not False)   # True

# Real example
is_raining = False
can_go_outside = not is_raining
print(f"\nIs it raining? {is_raining}")
print(f"Can I go outside? {can_go_outside}")
```

```python
# Combining all logical operators
username = "admin"
password = "secure123"
is_banned = False

correct_credentials = (username == "admin") and (password == "secure123")
can_login = correct_credentials and not is_banned

print(f"Correct credentials: {correct_credentials}")
print(f"Account banned: {is_banned}")
print(f"Can login: {can_login}")
```

```python
# Operator precedence: not > and > or
# Use parentheses for clarity!

a, b, c = True, False, True

# Without parentheses — can be confusing!
result1 = a or b and c        # b and c evaluated first, then or a

# With parentheses — clear and explicit!
result2 = a or (b and c)      # same result, but much clearer
result3 = (a or b) and c      # different result!

print(f"a or b and c        = {result1}")
print(f"a or (b and c)      = {result2}")
print(f"(a or b) and c      = {result3}")
```

```python
#In Python, bool is a special type that behaves like an integer in arithmetic operations
True == 1
False == 0

# So when you multiply a string by a boolean value, Python first treats the boolean as an integer and then performs string repetition.
print(f"False * 'jack' = {False * 'jack'}")
print(f"0 * 'jack' = {0 * 'jack'}")

print(f"True * 'jack' = {True * 'jack'}")
print(f"1 * 'jack' = {1 * 'jack'}")
```

```python
# An interesting thing
print(3 * 'jack')
print(True + True)
print(False + True)
```

  <h2 style="color: #553c9a; margin-top: 0;">🎯 Exercise 3 — BMI Calculator</h2>
  <p><strong>Task:</strong> Build a BMI (Body Mass Index) calculator.</p>
  <p><strong>Formula:</strong> <code>BMI = weight (kg) / (height (m))²</code></p>
  <p><strong>Requirements:</strong></p>
  <ul style="line-height: 2;">
    <li>Ask for weight in kg and height in cm</li>
    <li>Convert height from cm to meters</li>
    <li>Calculate BMI and show it with 2 decimal places</li>
    <li>Use comparison operators to check: <code>BMI &lt; 18.5</code> → Underweight, <code>&lt; 25</code> → Normal, <code>&lt; 30</code> → Overweight, else → Obese</li>
  </ul>
</div>

```python
# Exercise 3: BMI Calculator

# Get user input
name = input("Enter your name: ")
weight = float(input("Enter your weight (kg): "))
height_cm = float(input("Enter your height (cm): "))

# Convert and calculate
height_m = height_cm / 100         # cm → meters
bmi = weight / (height_m ** 2)     # BMI formula

# Determine BMI category using comparison operators
is_underweight = bmi < 18.5
is_normal      = (bmi >= 18.5) and (bmi < 25.0)
is_overweight  = (bmi >= 25.0) and (bmi < 30.0)
is_obese       = bmi >= 30.0

# Determine category string
category = "Underweight" * is_underweight + "Normal weight" * is_normal + "Overweight" * is_overweight + "Obese" * is_obese

# Display results
print(f"\n🏥 BMI Results for {name}")
print("-" * 30)
print(f"Weight   : {weight} kg")
print(f"Height   : {height_cm} cm ({height_m:.2f} m)")
print(f"BMI      : {bmi:.2f}")
print(f"Category : {category}")
```
